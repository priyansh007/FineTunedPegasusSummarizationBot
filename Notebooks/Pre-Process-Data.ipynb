```python
import pandas as pd
import requests
import json
import os
import re
from bs4 import BeautifulSoup
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Read the JSON data from file
with open('/content/drive/MyDrive/NewsNebula/data.json', 'r') as file:
    lines = file.readlines()
```

```python
lines[0]
```

```python
# Parse each JSON object and store in a list
data = []
for line in lines:
    json_object = json.loads(line)
    data.append(json_object)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
urls = df['link'][:100]
```

```python
urls[0]
```

```python
def download_data(url, folder_path, count):
  response = requests.get(url)
  if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    text = ""
    for paragraph in soup.find_all('p'): #The <p> tag defines a paragraph in the webpages
        text += paragraph.text
    text = text.lower()
    text = re.sub(r'\s+',' ',text)
    print(text)
    # Create a folder to store the files if it doesn't exist
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    # Save the extracted text data in a file named "dataq.txt"
    file_path = os.path.join(folder_path, "file_"+str(count)+".txt")
    with open(file_path, 'w') as file:
        file.write(text)

    print("Text data saved successfully.")
```

```python
download_data(urls[0], "/content/outputs", 1)
```
